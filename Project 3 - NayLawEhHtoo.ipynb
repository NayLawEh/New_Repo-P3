```python
trajectory_number =  8 #filter for trajectory 8

data_subset = ngsim_data[ngsim_data['trajectory_number'] == trajectory_number] #creates new dataframe with only data matching trajectory 8

plt.figure(figsize = (12,6)) #chart size
sns.lineplot(data = data_subset, x = 'Time', y = 'leader_acc(m/s^2)', label = 'Leader Acceleration', color = 'green', linestyle = '--') #plot leader acceleration by time, dashed line
sns.lineplot(data = data_subset, x = 'Time', y = 'follower_acc(m/s^2)', label = 'Follower Acceleration', color = 'orange') #plot follower speed by time, orange line
plt.xlabel('Time (seconds)', fontsize = 12, fontweight = 'bold') #horizontal label, bold
plt.ylabel('Acceleration (m/s^2)', fontsize = 12, fontweight = 'bold') #vertical label, bold
plt.title('Vehicle Acceleration Comparison (Trajectory #8)', fontsize = 20, fontweight = 'bold') #Title of chart, bold 
plt.legend() #displays legend

plt.show() #displays the generated chart
```

```python
def idm_acceleration(v, s, delta_v, v0, s0, T, a, b, delta): #function that calculates a vehicle's acceleration based on several physical parameters
    
    s_star = s0 + v * T + (v * delta_v) / (2 * np.sqrt(a * b)) #Calculates the "desired gap" 
    
    s_star = max(s_star, s0) #Ensures the desired gap never falls below the minimum safe distance.
    
    acceleration = a * (1 - (v/v0)**delta - (s_star/s)**2) #Computes the final acceleration by balancing the urge to speed up toward v0 against the need to maintain the gap s

    return acceleration #Output result
```

```python
trajectory_number = 8 #filter for trajectory 8
data_subset = ngsim_data[ngsim_data['trajectory_number'] == trajectory_number].reset_index(drop = True) #Filters for trajectory #8 and resets the row numbering to start from 0

time_data = data_subset['Time'].values #Converts the 'Time' column into a fast-processing NumPy array.
leader_position = data_subset['leader_position(m)'].values #Extracts the lead vehicle's distance markers as a NumPy array.
leader_speed = data_subset['leader_speed(m/s)'].values #Extracts the lead vehicle's speed values as a NumPy array.

dt = 0.1 # seconds
```

```python
v0 = 30    # desired velocity (m/s)
s0 = 2     # minimum spacing (m)
T = 1.5    # desired time headway (s)
a = 1.0    # maximum acceleration (m/s^2)  
b = 1.5    # comfortable deceleration (m/s^2)
delta = 4  # acceleration exponent

n_steps = len(time_data)        # number of simulation steps it will take, which will be equal to the length of the trajectory data
sim_position = np.zeros(n_steps)        # simulated follower position; I will create a vector with all zeroes  as a start
sim_speed = np.zeros(n_steps)        # simulated follower speed; I will create a vector with all zeroes  as a start
sim_acc = np.zeros(n_steps)        # simulated follower acceleration; I will create a vector with all zeroes  as a start

sim_position[0] = data_subset['follower_position(m)'].values[0]        #this makes the starting value [0] same as the follower vehicle position
sim_speed[0] = data_subset['follower_speed(m/s)'].values[0]        #Initializes the simulation with the real-world starting position and speed
```

```python
for i in range(n_steps - 1): # we do -1 since the first element is known (the step above) 

    gap = max(leader_position[i] - sim_position[i], 0,1) # Calculate the gap and speed difference at this time step
    delta_v = sim_speed[i] - leader_speed[i] #Calculates the relative speed difference between the follower and the leader.
    
    sim_acc[i] = idm_acceleration(sim_speed[i], gap, delta_v, v0, s0, T, a, b, delta) #Use the IDM function to calculate acceleration

    sim_speed[i+1] = max(sim_speed[i] + sim_acc[i]* dt, 0) #Update speed and position for the next time step
    sim_position[i+1] = sim_position[i] + sim_speed[i]* dt #Updates the vehicle's position based on its current speed and the time elapsed

print("Simulation complete!") #prints Simulation complete
```

```python
plt.figure(figsize = (12,6)) #chart size
plt.plot(time_data, leader_position, label = 'Leader (NGSIM)', color = "black", linewidth = 2) #plot leader position by time, black line
plt.plot(time_data, data_subset['follower_position(m)'].values, label = 'Follower (NGSIM)', color = "blue") #plot follower position by time, blue line
plt.plot(time_data, sim_position, label = 'Simulated Follower', color = "red", linestyle = '--') #plot simulated follower y time, red dashed line
plt.xlabel('Time (seconds)', fontsize = 12, fontweight = 'bold') #horizontal label, bold
plt.ylabel('Position (m)', fontsize = 12, fontweight = 'bold') #vertical label, bold
plt.title('Validation: Comparison of Real-World vs. Simulated Trajectories', fontsize = 20, fontweight = 'bold') #Title of chart, bold 
plt.legend() #plots legend
plt.grid(True, alpha = 0.3) #plots grid wih 30% transparency
plt.show() #displays the generated chart
```
